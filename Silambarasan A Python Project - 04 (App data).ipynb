### Q.No.: 9  . The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into suitable data type. (hint: Replace M with 1 million and K with 1 thousand, and drop the entries where size='Varies with device')



```python
df
```

```python
df.shape
```

```python
df['Size'].unique()
```

```python
df['Size']=='Varies with device'
```

```python
df[df['Size']=='Varies with device']
```

```python
df[df['Size']=='Varies with device'].index
```

```python
df.drop(df[df['Size']=='Varies with device'].index, inplace = True)
```

```python
df['Size'].unique()
```

```python
def convert_an(y):
    if type(y) == str:
    
        
        if 'M' in y:
            m = float(y.replace("M", ""))
            return m*1000000
        elif 'k' in y:
            k = float(y.replace('k', ''))
            return k*1000
        else:
            k= float(y.replace("K", ""))
            return k*1000
    else:
        return y
```

```python
df['Size'] = df['Size'].apply(convert_an)
```
